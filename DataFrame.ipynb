```python
# performing row wise .apply
points=pd.DataFrame({
    '1st point':[(3,4),(4,3),(0,1)],
    '2nd point':[(0,0),[2,3],[1,0]]
})
points
```

```python
#points['Distance']=points.apply(([points['1st point']][0]-[points['2nd point']][0])**2 + ([points['1st point']][0]-[points['2nd point']][0])**2)**0.5
```

```python
def distance(row):
    point_A=row['1st point']
    point_B=row['2nd point']
    return ((point_A[0]-point_B[0])**2+(point_A[1]-point_B[1])**2)**0.5
```

```python
points['Distance']=points.apply(distance,axis=1).round
points
```
